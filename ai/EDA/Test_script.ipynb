```python
import torch
import os
```

```python
# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5m')
```

```python
os.getcwd()
```

```python
# Images
dir = 'https://github.com/ultralytics/yolov5/raw/master/data/images/'
imgs = [dir + f for f in ('zidane.jpg', 'bus.jpg')]  # batch of images
```

```python
# Inference
results = model(imgs)
results.show()  # or .show(), .save()
```
